## Investigation Overview


> My main goal of this investigation is to look at how some variables affect prosperScore, the main focus is on 4 variables which are IsBorrowerHomeowner,EmploymentStatus, IncomeVerifiable, OnTimeProsperPayments. fromm the dat description, ProsperScore is the analysis of the risk given loans to individual based on past data.

## Dataset Overview

> This is a Loan data set which contains 113,937 loan records with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income,ProsperScore, EmploymentStatus and many others.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
prosperLoanData = pd.read_csv('prosperLoanData.csv')
```

```python
#function to plot bar chart
def barChart(Dataframe,variable, x_label, title):
    #Add figure size
    plt.figure(figsize = [14, 6])
    base_color = sb.color_palette()[3]
    sb.countplot(data=Dataframe, x=variable, color  = base_color );
    plt.xlabel(x_label)
    plt.title(title, fontsize=14)
    plt.show()
```

## Distribution of ProsperScore

> ProsperScore takes values from 1 to 10, this is a categorical variable with 1 been the most risk, this data is as of July 2009, from the statistics below we can see that majority of borrowers have a prosperScore of 4, 6 and 8 with 4 having the greatest count

```python
barChart(prosperLoanData,"ProsperScore","ProsperScore","Distribution of ProsperScore", )
```

## Distribution of Employment Status

> Employment status is one of the factors that will determine the if the prosperScore of a particular borrower will be high, we could see this relationship from the graph below.

```python
freq = prosperLoanData['EmploymentStatus'].value_counts()
order = freq.index
#Add figure size
plt.figure(figsize = [14, 6])
sb.countplot(data=prosperLoanData, x='EmploymentStatus', color=sb.color_palette()[2], order=order);
plt.xticks(rotation=90);
plt.title("Distribution of Employment Status", fontsize=14);
```

## IsBorrowerHomeowner vs ProsperScore
> Here i would like us to depict the relationship between IsBorrowerHomeowner and ProsperScore,how the employment status of an individual will affect their prosper rating, but this is nor a sufficient enough condition to determine prosper rating. The next slide will show a deeper relationship between these variables.



```python
#Add figure size
plt.figure(figsize = [14, 6])
sb.countplot(data=prosperLoanData, x= 'ProsperScore', hue='IsBorrowerHomeowner');
plt.title('Distribution of IsBorrowerHomeowner vs ProsperScore')
plt.legend(loc='upper right')
plt.show()
```

## Employment Status and ProsperScore by IsBorrowerHomeowner
> Here, we can see that not all borrowers with higher prosperRatings are necessarily employed, some of them are retired but still have a higher prosperScore becaue they own a home.


```python
#Add figure size
plt.figure(figsize = [14, 6])
ax = sb.pointplot(data = prosperLoanData, x = 'EmploymentStatus', y = 'ProsperScore', hue = 'IsBorrowerHomeowner',
                  dodge = 0.3, linestyles = "")
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90);
plt.title("Employment Status and ProsperScore by IsBorrowerHomeowner");
plt.legend(loc='upper left')
```

## Employment Status and ProsperScore by IncomeVerifiable
> Here, we can see that not all borrowers with higher prosperRatings are necessarily employed

```python
#Add figure size
plt.figure(figsize = [14, 6])
ax = sb.pointplot(data = prosperLoanData, x = 'EmploymentStatus', y = 'ProsperScore', hue = 'IncomeVerifiable',
                  dodge = 0.3, linestyles = "")
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90);
plt.title("Employment Status and ProsperScore by IncomeVerifiable");
plt.legend(loc='upper left')
```
